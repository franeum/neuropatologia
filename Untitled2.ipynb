# get values of blackscreens

## filter rows with blackscreen word

```python
import re

with open('provuma.txt','r') as f:
    data = f.readlines()

pat = re.compile(r'\[blackdetect')
blacks = [ x.strip(None) for x in data if re.search(pat, x) ]
blacks
```

## get values

```python
pattern = re.compile(r'(black_start:\d*\.\d*).*(black_end:\d*\.\d*).*(black_duration:\d*\.\d*)')
    
def get_value(s):
    splitted = s.strip(None).split(':')
    return float(splitted[1])

blackscreens_data = []

for black in blacks:
    s = re.search(pattern, black)
    if s:
        subdata = {}
        subdata['Start Time (seconds)'] = get_value(s.group(1))
        subdata['End Time (seconds)'] = get_value(s.group(2))
        subdata['Length (seconds)'] = get_value(s.group(3))
        blackscreens_data.append(subdata)
        
blackscreens_data
```
